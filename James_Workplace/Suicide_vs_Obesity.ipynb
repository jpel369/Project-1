```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
state_pop_df = pd.read_csv('Resources/state_population.csv')
state_suicide_df = pd.read_csv('Resources/state_suicide_rate.csv')
state_obesity_df = pd.read_csv('Resources/state_obesity_rate.csv')
```

```python
# Clean CSV
state_pop_df.rename(columns={'NAME':'STATE'}, inplace=True)
state_pop_df.set_index(state_pop_df['STATE'], inplace=True)
state_pop_df.drop(columns=['STATE'], inplace=True)
state_pop_df
```

```python
# Clean CSV
state_suicide_df.set_index(state_suicide_df['STATE'], inplace=True)
state_suicide_df.drop(columns=['STATE.1'], inplace=True)
state_suicide_df.drop(columns=['STATE'], inplace=True)
state_suicide_df
```

```python
# Clean Up CSV
state_obesity_df.set_index(state_obesity_df['STATE'], inplace=True)
state_obesity_df.drop(columns=['STATE.1'], inplace=True)
state_obesity_df.drop(columns=['STATE'], inplace=True)
state_obesity_df
# Create Final CSV for Suicide vs. Obesity Analysis
suicide_vs_obesity_df = state_pop_df.join(state_suicide_df)
suicide_vs_obesity_df = suicide_vs_obesity_df.join(state_obesity_df)
suicide_vs_obesity_df.to_csv('Suicide_vs_Obesity_US_States.csv', index=True)
```

```python
# Clean DataFrame
suicide_vs_obesity_df.dropna(inplace=True)
suicide_vs_obesity_df.rename(columns={'DEATHS':'NO_SUICIDES', 'PERCENTAGE':'OBESE PERCENTAGE'}, inplace=True)
suicide_vs_obesity_df.drop(columns=['RATE'], inplace=True)
suicide_vs_obesity_df['YEAR'] = suicide_vs_obesity_df['YEAR'].astype(int)
```

```python
suicide_vs_obesity_df
```

```python
# Calculate Percentage of Suicides per State
suicide_vs_obesity_df['NO_SUICIDES'] = suicide_vs_obesity_df['NO_SUICIDES'].str.replace(',','')
suicide_vs_obesity_df['NO_SUICIDES'] = suicide_vs_obesity_df['NO_SUICIDES'].astype(int)
suicide_vs_obesity_df['SUICIDE PERCENTAGE'] = (suicide_vs_obesity_df['NO_SUICIDES']/suicide_vs_obesity_df['POPULATION'])
```

```python
suicide_percentage = suicide_vs_obesity_df['SUICIDE PERCENTAGE'].apply(lambda x: x *100).round(3)
```

```python
suicide_frame_percent = suicide_percentage.to_frame()
suicide_frame_percent.rename(columns={'SUICIDE PERCENTAGE':'Percentage of Suicide'}, inplace=True)
```

```python
suicide_vs_obesity_df = suicide_vs_obesity_df.join(suicide_frame_percent, on='STATE')
suicide_vs_obesity_df.drop(columns=['SUICIDE PERCENTAGE'], inplace=True)
```

```python
suicide_vs_obesity_df.rename(columns={'Percentage of Suicide':'SUICIDE PERCENTAGE'}, inplace=True)
suicide_vs_obesity_df
```

```python
import matplotlib.pyplot as plt
plt.scatter(y=suicide_vs_obesity_df['OBESE PERCENTAGE'], x=suicide_vs_obesity_df['SUICIDE PERCENTAGE'], color='skyblue', alpha=0.5, edgecolor='black')
```
